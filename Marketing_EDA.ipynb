```python
#Perform EDA
#Visualization 1: Deposit Distributionimport seaborn as sns
#Compare customers who subscribed vs. did not subscribe.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('/content/bank.csv')
sns.countplot(x='deposit', data=df)
plt.title('Deposit Subscription Distribution')
plt.show()
```

```python
#Deposit by Job
#Identify professions with higher subscription rates.
plt.figure(figsize=(12,6))
sns.countplot(y='job', hue='deposit', data=df)
plt.title('Deposit Subscription by Job')
plt.show()
```

```python
#Deposit by Education
sns.countplot(x='education', hue='deposit', data=df)
plt.title('Deposit by Education Level')
plt.show()
```

```python
#Age Distribution
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
#Average Balance by Deposit
sns.boxplot(x='deposit', y='balance', data=df)
plt.title('Balance vs Deposit')
plt.show()
```

```python
#Contact Method Analysis
sns.countplot(x='contact', hue='deposit', data=df)
plt.title('Contact Method vs Deposit')
plt.show()
```
